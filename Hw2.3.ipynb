```python
import pandas as pd
from google.colab import drive
import matplotlib.pyplot as plt
drive.mount('/content/drive')
file_path = '/content/drive/My Drive/Colab Notebooks/bestsellers with categories.csv'
df = pd.read_csv(file_path)
print(df.head())
print('\n')
rows, columns = df.shape
print(f"Кількість рядків: {rows}")
print(f"Кількість стовпців: {columns}")
print('\n')
print("Про скільки книг зберігає дані датасет? \nПро 50 книг.")
print('\n')
df.columns = ['name', 'author', 'user_rating', 'reviews', 'price', 'year', 'genre'] # Список нових імен для колонок
missing_values = df.isna().sum()
print(missing_values)
print('\n')
print("Чи є в якихось змінних пропуски? \nНі.")
print('\n')
unique_genres = df['genre'].unique()
print(unique_genres)
print('\n')
print("Які є унікальні жанри? \nNon Fiction, Fiction.")
print('\n')
df['price'].hist(bins=25, edgecolor='black')
plt.title('Розподіл цін')
plt.xlabel('Ціна')
plt.ylabel('Кількість')
plt.show()
print('\n')
max_price = df['price'].max()
min_price = df['price'].min()
mean_price = df['price'].mean()
median_price = df['price'].median()
print(f"Максимальна ціна: {max_price}")
print(f"Мінімальна ціна: {min_price}")
print(f"Середня ціна: {mean_price}")
print(f"Медіанна ціна: {median_price}")
print('\n')
df_unique = df.drop_duplicates(subset='name', keep='first') #прибираємо повторення книг, залишаєм тільки перше згадування
highest_rating = df_unique['user_rating'].max()
print(f"Найвищий рейтинг у датасеті: {highest_rating}")
count_highest_rating = df_unique[df_unique['user_rating'] == highest_rating].shape[0]
print(f"Кількість книг з найбільшим рейтингом: {count_highest_rating}")
max_reviews_book = df.loc[df['reviews'].idxmax()]
print(f"Найбільше відгуків має: {max_reviews_book['name']}")
df_2015 = df[df['year'] == 2015]
most_expensive_book = df_2015.loc[df_2015['price'].idxmax()]
print(f"Найдорожча книга з Топ-50 у 2015 році: {most_expensive_book['name']}, її вартість: {most_expensive_book['price']}")
df_2010_fiction = df[(df['year'] == 2010) & (df['genre'] == 'Fiction')]
count_fiction_2010 = df_2010_fiction.shape[0]
print(f"Кількість книг жанру Fiction, що потрапили до Топ-50 у 2010 році: {count_fiction_2010}")
df_filtered = df[(df['user_rating'] == 4.9) & (df['year'].isin([2010, 2011]))]
count_books = df_filtered.shape[0]
print(f"Кількість книг з рейтингом 4.9, що потрапили до рейтингу у 2010 та 2011 роках: {count_books}")
df_filtered_by_price = df[(df['year'] == 2015) & (df['price'] < 8)]
df_sorted_by_price = df_filtered_by_price.sort_values(by='price')
print(df_sorted_by_price)
last_book = df_sorted_by_price.iloc[-1]
print(f"Остання книга у відсортованому списку: {last_book['name']}, її ціна: {last_book['price']}")

# Підрахунок кількості книг за жанрами
genre_counts = df['genre'].value_counts()

plt.figure(figsize=(7,7))
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', startangle=140, colors=['#66b3ff','#ff9999'])
plt.title('Розподіл книг за жанрами')
plt.axis('equal')  # Щоб пиріг був круглим
plt.show()


```
